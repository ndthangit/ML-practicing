## setup data

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline

loan_data = pd.read_csv("data/loan_data.csv")
print(loan_data.shape)
loan_data.head(20)
```

## explore

```python
loan_data.describe()
```

```python
average_fico= (loan_data.loc[loan_data['credit.policy'] == 1, 'fico'].sum())/((loan_data['credit.policy'] == 1).count())
print("Average fico score of customer meeting the credit underwriting criteria: ",average_fico)
```

```python
summary_loan_data = loan_data.drop(['log.annual.inc', 'revol.bal', 'revol.util', 'inq.last.6mths'], axis=1)
summary_loan_data.head()
```

```python
purpose_of_loan = loan_data['purpose'].value_counts()
print(purpose_of_loan)
```

```python
statement = loan_data['not.fully.paid'].value_counts()
print(statement)
```

## visualization


```python
fig, (table1, table2) = plt.subplots(nrows=2, ncols=1, figsize=(10,10))

# table 1
statement.plot(kind='bar', ax=table1)
table1.set_xlabel('Statement')
table1.set_ylabel('Quantity')
table1.set_xticklabels(table1.get_xticklabels(), rotation=360)
table1.set_title('The statistic of the quantity of people not fully paid')

# add exactly statistic
for i, value in enumerate(statement):
    table1.text(i, value, str(value), ha='center', va='bottom')



#table 2
purpose_of_loan.plot(kind='barh', ax=table2)
table2.set_ylabel('Purpose')
table2.set_xlabel('Quantity')
table2.set_title('The statistic of the purpose of customers')
# add exactly statistic 
for i, value in enumerate(purpose_of_loan):
    table2.text(value, i, str(value), ha='left', va='center')

```
